```python
moviemat = df.pivot_table(index='userId',columns='title',values='rating')
moviemat.head()
```

```python
 PulpFiction_user_ratings= moviemat['Pulp Fiction (1994)']
SilenceoftheLambsThe_user_ratings = moviemat['Silence of the Lambs, The (1991)']
PulpFiction_user_ratings.head()
```

```python
similar_to_PulpFiction = moviemat.corrwith(PulpFiction_user_ratings)
similar_to_SilenceoftheLambsThe = moviemat.corrwith(SilenceoftheLambsThe_user_ratings)
```

```python
corr_PulpFiction = pd.DataFrame(similar_to_PulpFiction,columns=['Correlation'])
corr_PulpFiction.dropna(inplace=True)
corr_PulpFiction.head()
```

```python
corr_PulpFiction.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_PulpFiction= corr_PulpFiction.join(ratings['num of ratings'])
corr_PulpFiction.head()
```

```python
corr_PulpFiction[corr_PulpFiction['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_SilenceoftheLambsThe = pd.DataFrame(similar_to_SilenceoftheLambsThe,columns=['Correlation'])
corr_SilenceoftheLambsThe.dropna(inplace=True)
corr_SilenceoftheLambsThe = corr_SilenceoftheLambsThe.join(ratings['num of ratings'])
corr_SilenceoftheLambsThe[corr_SilenceoftheLambsThe['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
